```python
import fasttext
import fasttext.util
fasttext.util.download_model('en', if_exists='ignore')  # English
```

```python
ft = fasttext.load_model('cc.en.300.bin')
```

2\. How large is the vocabulary?

```python
len(ft.get_words())
```

3\. How many dimensions does the model have?

```python
ft.get_dimension()
```

4\. Print the word vector for "Munich".

```python
ft.get_word_vector("Munich")
```

5\. What are the most similar terms to "LRZ"?

```python
ft.get_nearest_neighbors("LRZ", k=10)
```

6\. Find the analogy: Berlin is to Germany as ??? is to Denmark.

```python
ft.get_analogies("Berlin", "Germany", "Denmark", k=1)
```
